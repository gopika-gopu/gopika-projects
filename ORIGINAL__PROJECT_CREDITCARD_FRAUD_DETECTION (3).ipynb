```python
# checking whether it is balanced or not
```

```python
fraud_count = data["is_fraud"].value_counts()[1]  # Fraud cases
non_fraud_count = data["is_fraud"].value_counts()[0]  # Non-fraud cases

imbalance_ratio = non_fraud_count / fraud_count
print(f"Imbalance Ratio: {imbalance_ratio:.2f}")
```

so, the dataset is highly imbalanced. so we have to make it balanced

```python
from imblearn.over_sampling import SMOTE

# Define Features and Target
X = data.drop(["is_fraud","first","last","state"], axis=1)
y = data["is_fraud"]

# Apply SMOTE
smote = SMOTE(sampling_strategy=0.1, random_state=42)  # Adjust as needed
X_resampled, y_resampled = smote.fit_resample(X, y)

# Check new class distribution
print(y_resampled.value_counts())
```

```python
X.shape
```

spliting into x_train and x_test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X_resampled,y_resampled,test_size=0.2,random_state=42)
```
